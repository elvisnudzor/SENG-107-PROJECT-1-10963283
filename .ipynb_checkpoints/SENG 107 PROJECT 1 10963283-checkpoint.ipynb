### Data Statistics


```python
colours = ['red','orange','yellow','green','blue','beige','violet','black','pink','grey','brown','purple']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5,0, 0.5)
meanValues = [meanFA, meanVA, meanCA, meanRS, meanC, meanFSD, meanTSD, meanD, meanPH, meanS,meanA, meanQ]
meanTopics = ['FA',
             'VA',
             'CA',
             'RS',
             'C',
             'FSD',
             'TSD',
             'density',
             'pH',
             'sulphates',
             'alcohol',
             'quality']
```

```python
xtitle = 'Column names'
ytitle = 'Mean values'
legendlabels = ['fixed_acidity(FA)',
               'volatile_acidity(VA)',
               'citric_acid(CA)',
               'residual_sugar(RS)',
               'chlorides(C)',
               'free_sulfur_dioxide(FSD)',
               'total_sulfur_dioxide(TSD)',
               'density',
               'pH',
               'sulphates',
               'alcohol',
               'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

```python
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colours, labeldistance =1.2)
plt.axis('equal')
plt.legend(title='index', labels=legendlabels, loc='upper right',)
plt.title('Mean Values for each column in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
plt.bar(labels, sizes, width=0.5,color=colours)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('comparison of mean of each series in a bar chart')
plt.legend(loc='upper right')
plt.show()
```
